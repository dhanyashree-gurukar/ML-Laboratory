# Scatter-Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, c='blue', alpha=0.5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Scatter Plot")
plt.show()
```

# Box-Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np.random.rand(100)

sns.boxplot(data=data)
plt.title("Box Plot")
plt.show()
```

```python
#Using toyota dataset
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ToyotaCorolla.csv")

data["Price"] = pd.to_numeric(data["Price"], errors="coerce")
data["HP"] = pd.to_numeric(data["HP"], errors="coerce")
data["KM"] = pd.to_numeric(data["KM"], errors="coerce")
data_clean = data[["Price", "HP", "KM"]].dropna()

plt.boxplot([data_clean["Price"], data_clean["HP"], data_clean["KM"]])
plt.xticks([1, 2, 3], ["price", "HP", "KM"])
plt.title("Boxplot of Price, HP and KM")
plt.show()
```

# Heat-Map

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 12)

sns.heatmap(data, annot=True, cmap='viridis')
plt.title("Heat Map")
plt.show()
```
