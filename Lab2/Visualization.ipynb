# Scatter-Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, c='blue', alpha=0.5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Scatter Plot")
plt.show()
```

# Box-Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np.random.rand(100)

sns.boxplot(data=data)
plt.title("Box Plot")
plt.show()
```

```python
#Using toyota dataset
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ToyotaCorolla.csv")

data["Price"] = pd.to_numeric(data["Price"], errors="coerce")
data["HP"] = pd.to_numeric(data["HP"], errors="coerce")
data["KM"] = pd.to_numeric(data["KM"], errors="coerce")
data_clean = data[["Price", "HP", "KM"]].dropna()

plt.boxplot([data_clean["Price"], data_clean["HP"], data_clean["KM"]])
plt.xticks([1, 2, 3], ["price", "HP", "KM"])
plt.title("Boxplot of Price, HP and KM")
plt.show()
```

# Heat-Map

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 12)

sns.heatmap(data, annot=True, cmap='viridis')
plt.title("Heat Map")
plt.show()
```

```python
#Using toyota dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("ToyotaCorolla.csv")
sns.heatmap(data[["Price", "KM", "Doors", "Weight"]].corr(), cmap='jet')
plt.show()
```

# Contour-Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, levels=20, cmap='RdGy')
plt.title("Contour Plot")
plt.show()
```

```python
#Using Toyota dataset
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dataset = pd.read_csv("ToyotaCorolla.csv")
dataset.replace('??', pd.NA, inplace=True)
dataset["KM"] = pd.to_numeric(dataset["KM"], errors="coerce")
dataset["HP"] = pd.to_numeric(dataset["HP"], errors="coerce")
dataset["Price"] = pd.to_numeric(dataset["Price"], errors="coerce")

dataset_clean = dataset.dropna(subset=["KM", "HP", "Price"])

x = dataset_clean["KM"]
y = dataset_clean["HP"]
z = dataset_clean["Price"]

plt.tricontourf(x, y, z, levels=20, cmap='jet')
plt.colorbar(label="Price")
plt.xlabel("KM")
plt.ylabel("HP")
plt.title("Price distribution based on KM and HP")
plt.show()
```

# 3D Surface-Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot")
plt.show()
```

```python
#Using Toyota dataset
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("ToyotaCorolla.csv")
dataset.replace('??', pd.NA, inplace=True)
dataset["KM"] = pd.to_numeric(dataset["KM"], errors="coerce")
dataset["HP"] = pd.to_numeric(dataset["HP"], errors="coerce")
dataset["Price"] = pd.to_numeric(dataset["Price"], errors="coerce")

dataset_clean = dataset.dropna(subset=["KM", "HP", "Price"])

x = dataset_clean["KM"]
y = dataset_clean["HP"]
z = dataset_clean["Price"]

fig = plt.figure(figure = (10, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_trisurf(x, y, z, cmap='jet')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label="Price")

ax.set_xlabel("KM")
ax.set_ylabel("HP")
ax.set_zlabel("Price")
ax.set_title("3D Surface Plot of price based on KM and HP")
plt.show()
```
